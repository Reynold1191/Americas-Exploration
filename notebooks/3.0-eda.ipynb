**Question 01**: How does the correlation between features change in the top 10 countries with the LARGEST population?

When we have an answer for this question, we might see how features interaction with each other when the conutry population is large.

```python
# PRE-PROCESSING IF NEEDED
# YOUR CODE HERE (OPTION)
data_Q1 = cleaned_df.copy()

def top_10_popu_largest(df, year, top):
    df_year = df[df['Year'] == year]
    top_list = df_year.sort_values("Total Population", ascending=False).head(top)
    return top_list

top = top_10_popu_largest(data_Q1, 2021, 10)

top.drop(['Year', 'Country'], inplace=True, axis='columns')
top
```

```python
# YOUR CODE HERE
features_corr_10_largest = top.corr()

fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(features_corr_10_largest, cmap='RdBu', center=0,ax=ax)
plt.show()
```

```python
n_positive_corr = 5
top_positive_corr = features_corr_10_largest.stack().loc[features_corr_10_largest.stack() < 1.0].nlargest(n_positive_corr)

n_negative_corr = 5
top_negative_corr = features_corr_10_largest.stack().loc[features_corr_10_largest.stack() > -1.0].nsmallest(n_negative_corr)

print(top_positive_corr)
print(top_negative_corr)
```

**YOUR ANSWER**

Here are my observations:
- There are some 0 correlations (e.g. Fossil Fuel Consumption and Renewable Energy Consumption) --> Same value
- The most 3 positive correlations: Total Population - Female Population, Total Population - Male Population, Male Population - Female Population, which have the value of roughly 1 --> one increases so does the another
- The most 3 negative correlations: Female life expectancy - Birth rate, Compulsory Education Dur. - Male life expectancy, Birth Rate - Male life expectancy --> the increase of one leads to the decrease of another

## Making your own question 01? (1.5 points)

<font color=lightgreen>**Context**</font>: Male and Female 

<font color=lightgreen>**Question**</font>: What are the countries having the highest gap between male and female?

```python
# YOUR CODE HERE (OPTION)
# Male Population
# Female Population 
OQ1_avg_male_df = pd.DataFrame()
OQ1_avg_total_df = pd.DataFrame()
OQ1_avg_avg_df = pd.DataFrame()

OQ1_avg_male_df["Male Population"] = cleaned_df.groupby("Country")["Male Population"].mean()
OQ1_avg_total_df["Total Population"] = cleaned_df.groupby("Country")["Total Population"].mean()


OQ1_avg_avg_df["Country"] = OQ1_avg_male_df.index  

OQ1_avg_male_df.reset_index(drop=True, inplace=True)
OQ1_avg_total_df.reset_index(drop=True, inplace=True)


OQ1_avg_avg_df["Avg.Sex"] = (OQ1_avg_male_df["Male Population"] / OQ1_avg_total_df["Total Population"])

OQ1_avg_avg_df.sort_values("Avg.Sex", inplace=True, ascending=False)
```

## Making your own question 02? (1.5 points)

<font color=lightgreen>**Context**</font>: As we know, education plays a vital role in society. 

<font color=lightgreen>**Question**</font>: What is the level of each country?

**YOUR ANSWER HERE**

### Pre-processing

```python
# YOUR CODE HERE (OPTION)
# School enrollment, primary
# School enrollment, tertiary
# Primary completion rate
# Literacy rate
#OW2 means own question 2

OQ2_avg_primary_df = pd.DataFrame()
OQ2_avg_tertiary_df = pd.DataFrame()
OQ2_avg_completion_df = pd.DataFrame()
OQ2_avg_literacy_df = pd.DataFrame()
OQ2_avg_avg_df = pd.DataFrame()

OQ2_avg_primary_df["School enrollment, primary"] = cleaned_df.groupby("Country")["School enrollment, primary"].mean()
OQ2_avg_tertiary_df["School enrollment, tertiary"] = cleaned_df.groupby("Country")["School enrollment, tertiary"].mean()
OQ2_avg_completion_df["Primary completion rate"] = cleaned_df.groupby("Country")["Primary completion rate"].mean()
OQ2_avg_literacy_df["Literacy rate"] = cleaned_df.groupby("Country")["Literacy rate"].mean()

OQ2_avg_avg_df["Country"] = OQ2_avg_primary_df.index  

OQ2_avg_primary_df.reset_index(drop=True, inplace=True)
OQ2_avg_tertiary_df.reset_index(drop=True, inplace=True)
OQ2_avg_completion_df.reset_index(drop=True, inplace=True)
OQ2_avg_literacy_df.reset_index(drop=True, inplace=True)

OQ2_avg_avg_df["Avg.Education"] = (OQ2_avg_primary_df["School enrollment, primary"] + OQ2_avg_tertiary_df["School enrollment, tertiary"] 
                                   + OQ2_avg_completion_df["Primary completion rate"] + OQ2_avg_literacy_df["Literacy rate"]) / 4

OQ2_avg_avg_df.sort_values("Avg.Education", inplace=True, ascending=False)
```
